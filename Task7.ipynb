```python
from textwrap import fill
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataset is loaded into a DataFrame named ds
# For this example, let's create a sample dataset similar to the one we have the statistics for
np.random.seed(0)
ds = pd.DataFrame({
    'AGE': np.random.randint(21, 39, size=1000)  # Creating a sample 'AGE' column
})

# Extract the 'AGE' column
column_data = ds['AGE']

# Calculate statistics
mean = column_data.mean()
median = column_data.median()
mode = column_data.mode().iloc[0]
std_dev = column_data.std()
q1 = column_data.quantile(0.25)
q3 = column_data.quantile(0.75)

# Print the results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
print(f"25th percentile (Q1): {q1}")
print(f"75th percentile (Q3): {q3}")

# Create a figure
plt.figure(figsize=(8, 5))

# Create a boxplot using seaborn
sns.boxplot(x=column_data, color='skyblue')

# Customize the figure
plt.title('Age Distribution (Boxplot)')
plt.xlabel('Age')
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataset is loaded into a DataFrame named ds
# For this example, let's create a sample dataset similar to the one we have the statistics for
np.random.seed(0)
ds = pd.DataFrame({
    'AGE': np.random.randint(21, 39, size=1000)  # Creating a sample 'AGE' column
})

# Extract the 'AGE' column
age_data = ds['AGE']

# Calculate mean and standard deviation
mean_age = np.mean(age_data)
std_dev_age = np.std(age_data)

# Create histogram
sns.histplot(age_data, bins=20)

plt.ylabel('Frequency')
plt.xlabel('Age')
plt.title("Histogram of Ages")
sns.despine(offset=10, trim=True)

# Plot mean and standard deviation lines
plt.axvline(mean_age, color='green', linestyle='-', linewidth=2, label='Mean')
plt.axvline(mean_age + std_dev_age, color='red', linestyle='-', linewidth=2, label='Mean + SD')
plt.axvline(mean_age - std_dev_age, color='red', linestyle='-', linewidth=2, label='Mean - SD')

plt.legend()
plt.show()

```
